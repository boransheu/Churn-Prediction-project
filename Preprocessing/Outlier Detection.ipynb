```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/home/jupyter/Data_files/data_vfinal_03_17_23.csv")
```

```python
selected_columns = [  'months_elapsed',
       'purchase_times', 'lifetime_spent', 'lifetime_quantity',
       'per_order_spent', 'per_order_quantity', 'distinct_category_count', 'discount_frequency_order',
       'discount_frequency_product', 'product_added_count',
       'product_added_to_wishlist_count', 'product_removed_count',
       'customizer_started_count', 'customizer_completed_count',
       'cart_viewed_count', 'checkout_started_count', 'order_cancelled_count',
       'email_received', 'email_open_rate']
```

```python
corr = df[selected_columns].corr()
```

```python
sns.heatmap(corr, cmap='coolwarm', annot=True)

# set the size of the heatmap
sns.set(rc={'figure.figsize':(20,20)})
plt.show()
```
